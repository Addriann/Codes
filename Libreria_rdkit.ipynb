```python
from IPython.display import display
from rdkit import Chem
from rdkit.Chem import Draw

# Crear la molécula a partir de SMILES (simplified molecular-input line-entry system)
pet_monomer = Chem.MolFromSmiles("O=C(O)Cc1ccc(C(=O)OCCOC(=O)c2ccc(C(=O)O)cc2)cc1")

# Dibujo
img = Draw.MolToImage(pet_monomer, size=(1024, 768))
display(img)
```
